```r
library("ggplot2")
```

function are imported without a namespace:  
`qplot` from `ggplot2` just gets imported as `qplot`  

BTW, qplot let's you plot quick simple plots.

```r
qplot(c(1,2,3), c(4,5,6))
```

```r
x <- c(-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1)
y <- x^3
qplot(x,y)
```

```r
qplot(c(1, 2, 2, 2, 3, 3), binwidth=1)
```
